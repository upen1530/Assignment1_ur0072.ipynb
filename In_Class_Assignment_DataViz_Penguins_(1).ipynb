## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
# Create the scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
                hue='species', palette='Set2', s=70, edgecolor='black')
plt.title('Penguin Bill Length vs Bill Depth by Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.legend(title='Species')
plt.show()


```

**Your interpretation (2–3 sentences):**

_Write your observations here._

```python
Gentoo penguins usually have longer bills but shallower depths.
Adelie penguins tend to have shorter, deeper bills.
Chinstrap penguins fall between the two, showing moderate bill length and depth.
```

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: # Histogram with 10 bins
plt.figure(figsize=(7,5))
sns.histplot(penguins['bill_length_mm'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Bill Length (10 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')
plt.show()
# Histogram with 10 bins
```

```python
# Histogram with 30 bins
plt.figure(figsize=(7,5))
sns.histplot(penguins['bill_length_mm'], bins=30, color='salmon', edgecolor='black')
plt.title('Histogram of Bill Length (30 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')
plt.show()
```

**Your comparison (1–2 sentences):**

_Write your observations here._

```python
With 10 bins, the histogram looks smoother and highlights general trends.
With 30 bins, more detail appears — small peaks and dips become visible, but the plot looks noisier.
```

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species

# Boxplot of body mass by species
plt.figure(figsize=(8,6))
sns.boxplot(data=penguins, x='species', y='body_mass_g', palette='pastel')
plt.title('Body Mass of Penguin Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```

**Your summary (1–2 sentences):**

_Write your observations here._

```python
Gentoo penguins are the heaviest on average, while Adelie penguins are the lightest.
Chinstrap penguins fall in between, showing moderate body weights.
```
