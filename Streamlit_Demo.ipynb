```python
# Create app.py - Part 1 (Imports & Config)

%%writefile app.py
import streamlit as st
import pandas as pd

# Set page configuration
st.set_page_config(page_title="Student Grade Tracker", layout="wide")
```

**Explanation:**

- `%%writefile app.py` â†’ Creates a new file called **app.py** in Colab  
- `import streamlit as st` and `import pandas as pd` â†’ Import **Streamlit** and **Pandas**  
- `st.set_page_config()` â†’ Sets up the **page layout and title** (must be the *first* Streamlit command!)  
- **Note:** `%%writefile` **overwrites** the file â€” use it only when creating or replacing the file.


```python
# Add Title & Description (Append to app.py)

%%writefile -a app.py
# Title and description
st.title("ðŸ“š Student Grade Tracker")
st.write("A simple web application built with Streamlit to view and filter student grades.")
```

**Explanation:**

- `%%writefile -a app.py` â†’ The `-a` flag **appends** to the existing file (instead of overwriting)  
- `st.title()` â†’ Creates a **large heading**, often with emojis for visual appeal  
- `st.write()` â†’ Displays **text, data, or Markdown** on the Streamlit app  


```python
# Add Sample Data (Append to app.py)

%%writefile -a app.py
# Sample data - in-memory storage
students_data = {
    "Name": ["Alice Johnson", "Bob Smith", "Carol White", "David Lee", "Emma Davis"],
    "Math": [92, 85, 88, 76, 95],
    "Science": [88, 90, 85, 82, 92],
    "English": [85, 88, 92, 88, 90],
    "History": [90, 82, 88, 85, 88]
}

df = pd.DataFrame(students_data)
```

**Explanation:**  
- We create a dictionary with student names and grades across 4 subjects â†’ Creates the **raw data** for the app  
- `pd.DataFrame()` â†’ Converts the dictionary into a **table-like structure (2D data)**  
- This is the foundational data that our app will work with â†’ Serves as the **core dataset** for the app  
- In real projects, you'd load from CSV/database, but here we hardcode for learning â†’ Demonstrates a **simplified example**; normally data comes from **external sources**


```python
# Add Sidebar Filter (Append to app.py)

%%writefile -a app.py
# Sidebar filters
st.sidebar.header("Filter Options")
selected_student = st.sidebar.selectbox("Select a student:", df["Name"])
```

**Explanation:**  
- `st.sidebar` â†’ Creates a **control panel** on the left side of the app  
- `st.sidebar.header()` â†’ Adds a **heading** in the sidebar  
- `st.sidebar.selectbox()` â†’ Creates a **dropdown menu** for student selection  
- The selected student is stored in `selected_student` variable â†’ Stores the **user's selection** for use in the app  
- **Interactivity** â†’ When a user selects a student, **Streamlit re-runs the entire script** to update the app dynamically


```python
# Display Selected Student's Grades (Append to app.py)

%%writefile -a app.py
# Display grades for selected student
st.subheader(f"Grades for {selected_student}")
student_row = df[df["Name"] == selected_student].iloc[0]

# Create columns for subject grades
col1, col2, col3, col4 = st.columns(4)
with col1:
    st.metric("Math", student_row["Math"])
with col2:
    st.metric("Science", student_row["Science"])
with col3:
    st.metric("English", student_row["English"])
with col4:
    st.metric("History", student_row["History"])
```

**Explanation:**  
- `st.subheader()` â†’ Creates a **smaller heading**  
- `df[df["Name"] == selected_student]` â†’ Boolean filtering to get the **selected student's row**  
- `.iloc[0]` â†’ Gets the **first row as a Series** (all grades for that student)  
- `st.columns(4)` â†’ Splits the screen into **4 equal columns**  
- `st.metric()` â†’ Displays **numeric values with labels** (great for KPIs)  
- **Layout Design** â†’ Arranging metrics **horizontally** is cleaner than vertical stacking


```python
# Calculate & Display Average (Append to app.py)

%%writefile -a app.py
# Calculate and display average grade
subjects = ["Math", "Science", "English", "History"]
average = student_row[subjects].mean()
st.write(f"**Average Grade: {average:.1f}**")
```

**Explanation:**  
- `subjects` â†’ List of **column names** we want to average  
- `student_row[subjects]` â†’ Selects **only these columns** from the student's data  
- `.mean()` â†’ Pandas method to calculate the **arithmetic average**  
- `f"**...**"` â†’ F-string with **Markdown bold formatting**  
- `:.1f` â†’ Formats the number to **1 decimal place**  
- **Data Science Skill** â†’ **Aggregation and summary statistics** are fundamental!


```python
# Display All Students Table (Append to app.py)

%%writefile -a app.py
# Show all students in a table
st.subheader("All Students")
st.dataframe(df, use_container_width=True)
```

**Explanation:**  
- `st.dataframe()` â†’ Displays the **entire DataFrame** as an interactive table  
- Users can sort columns, search, and scroll through data â†’ Provides built-in interactivity for the table  
- `use_container_width=True` â†’ Makes the table **stretch to fill available width**  
- **Interactive Tables** â†’ Streamlit handles all the **HTML/CSS automatically**


```python
# Add Grade Calculator (Append to app.py)

%%writefile -a app.py
# Add a simple grade calculator
st.subheader("Grade Calculator")
grades = st.slider("Enter test grades:", 0, 100, (50, 90))
calculated_average = (grades[0] + grades[1]) / 2
st.info(f"Average of {grades[0]} and {grades[1]}: **{calculated_average:.1f}**")
```

**Explanation:**  
- `st.slider()` â†’ Creates a **slider widget** for numeric input  
- Arguments: (label, min, max, default_values) â†’ Defines the **slider parameters**  
- `(50, 90)` â†’ Default values create a **range slider** (two values)  
- `grades[0]` and `grades[1]` â†’ Access the **two selected values** from the slider  
- `st.info()` â†’ Displays an **informational message** in a blue box  
- **Reactivity** â†’ Moving the slider **instantly recalculates and updates** the app
